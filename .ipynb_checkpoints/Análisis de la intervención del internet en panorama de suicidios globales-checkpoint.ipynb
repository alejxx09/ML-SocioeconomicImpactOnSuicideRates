# Análisis de la intervención del internet en panorama de suicidios globales
___

## Introducción

En este estudio, trabajamos con dos datasets para responder a la pregunta: **¿El internet afecta de alguna manera al panorama de suicidios en el mundo?**

## Datasets Utilizados

### 1. Tasas de Suicidio (1990-2022)

El primer dataset que analizamos es `suicide_rates_1990-2022.csv`. Este dataset ofrece una visión integral de las tendencias globales de suicidio desde 1990 hasta 2022, enfocándose en las tasas de suicidio crudas. Los datos proporcionan información sobre la región, el país, el año y las principales métricas de suicidio, como:

- **Suicide Counts**: El número total de suicidios registrados.
- **Cause-Specific Death Percentage**: El porcentaje de muertes totales atribuidas al suicidio. Esto se calcula tomando el total de muertes por suicidio dividido por las muertes causadas por otras causas, incluido el suicidio, y multiplicando por 100.
- **Death Rate per 100,000 Population (Crude)**: El número no ajustado de suicidios por cada 100,000 personas en la población.

### 2. Uso del Internet

El segundo dataset, `share-of-individuals-using-the-internet.csv`, nos proporciona el porcentaje de la población que utiliza internet durante el año correspondiente.

## Objetivo del Estudio

Con estos datos, podemos observar de manera preliminar si existe o no una correlación entre el uso de internet y las tasas de suicidio globales.

## Manejo de Datos

### Importacion de librerias

Primero, importamos las librerías necesarias para el análisis. Estas incluyen `pandas` para la manipulación de datos, `matplotlib.pyplot` y `seaborn` para la visualización de datos, `numpy` para operaciones numéricas, y `LinearRegression` de `sklearn` para el análisis de regresión lineal.


```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression


# Cargar los datos
uso_internet = pd.read_csv('individuos_uso_internet.csv')
data_suicidios = pd.read_csv('edad_suicidios_1990_2022.csv')
```

### Manipulación de los datos

Vamos a manipular los datos para poder usarlos de manera correcta, debemos realizar la limpieza de los outliers, generar datos cuando sea necesario y la creación de columnas calculadas.

```python
# Evaluamos los valores del dataset
uso_internet.head()
```

```python
# Vamos a evaluar la cantidad de datos que tenemos por año
columna = 'Year'

# Crear el gráfico
plt.figure(figsize=(10, 6))
sns.countplot(x=columna, data=uso_internet)
plt.title('Conteo de valores para los año')
plt.xticks(rotation=30)
plt.show()
```

### Transformación y Regresión de Datos del Uso de Internet

#### Conversión de Tipo de Datos

Primero, nos aseguramos de que la columna `Year` del DataFrame `uso_internet` esté en formato entero. Esto es importante para realizar cálculos precisos y manipulaciones de datos.

#### Creación de Nueva Columna

Creamos una nueva columna `Year_numeric` que representa los años en formato numérico restando el año mínimo del conjunto de datos al año correspondiente. Esto facilita el uso de técnicas de regresión lineal.

#### Inicialización de Almacenamiento de Nuevos Datos

Inicializamos una lista `new_rows` para almacenar los datos generados que no están presentes en el dataset original.

#### Iteración Sobre Países

Iteramos sobre cada país presente en el DataFrame `uso_internet`:

- Filtramos los datos de cada país.
- Seleccionamos solo los años con valores de uso de internet diferentes de 0 para la regresión.
- Si no hay datos válidos para un país, omitimos ese país.

#### Preparación y Ajuste del Modelo de Regresión

Preparamos los datos filtrados para la regresión lineal:

- `X` contiene los valores numéricos de los años.
- `y` contiene los porcentajes de uso de internet correspondientes.

Ajustamos un modelo de regresión lineal a estos datos.

#### Generación de Nuevos Datos

Generamos nuevos datos para los años entre 1990 y 2022 que no están presentes en el dataset original:

- Calculamos los valores predecidos usando el modelo de regresión lineal.
- Almacenamos los nuevos datos en la lista `new_rows`.

```python

# Convertir 'Year' a tipo int
uso_internet['Year'] = uso_internet['Year'].astype(int)

# Crear una columna de características
uso_internet['Year_numeric'] = uso_internet['Year'] - uso_internet['Year'].min()

# Lista para almacenar nuevos datos
new_rows = []

# Iterar sobre cada país
for country in uso_internet['Entity'].unique():
    country_df = uso_internet[uso_internet['Entity'] == country]
    
    # Filtrar los años existentes que no son 0 para la regresión
    valid_df = country_df[country_df['Individuals using the Internet (% of population)'] != 0]
    if valid_df.empty:
        continue  # Si no hay datos válidos, omite el país
    
    # Preparar los datos para la regresión
    X = valid_df[['Year_numeric']].values
    y = valid_df['Individuals using the Internet (% of population)'].values
    
    model = LinearRegression()
    model.fit(X, y)
    
    # Generar nuevos datos para años entre 1990 y 2022 si no están presentes
    years_to_add = set(range(1990, 2023)) - set(country_df['Year'])
    for year in years_to_add:
        X_new = np.array([[year - uso_internet['Year'].min()]])
        predicted_value = model.predict(X_new)[0]
        new_rows.append({'Entity': country, 'Code': country_df['Code'].values[0], 'Year': year, 
                        'Individuals using the Internet (% of population)': predicted_value})

# Crear DataFrame con los nuevos datos
new_data_df = pd.DataFrame(new_rows)

# Agregar los nuevos datos al DataFrame original
uso_internet = pd.concat([uso_internet, new_data_df], ignore_index=True)

# Asegurarse de que los valores de 0 en el DataFrame original se mantengan
uso_internet['Individuals using the Internet (% of population)'] = uso_internet.apply(
    lambda row: 0 if row['Individuals using the Internet (% of population)'] == 0 else row['Individuals using the Internet (% of population)'],
    axis=1
)

# Opcionalmente, ordenar el DataFrame por país y año
uso_internet.sort_values(by=['Entity', 'Year'], inplace=True)
uso_internet
```

```python
# Podemos observar con esta gráfica que nuestra data se ve mejor equilibrada.
columna = 'Year'

# Crear el gráfico
plt.figure(figsize=(10, 6))
sns.countplot(x=columna, data=uso_internet)
plt.title('Conteo de valores para los año')
plt.xticks(rotation=30)
plt.show()
```

```python
# Renombrar los encabezados de Uso Internet
uso_internet.rename(columns={
    'Entity': 'Pais',
    'Code': 'Codigo Pais',
    'Year': 'Year',
    'Individuals using the Internet (% of population)': 'PoblacionUsandoInternetPorcentaje'
}, inplace=True)
uso_internet.head()
```

```python
data_suicidios.head()
```

```python
# Renombrar los encabezados de la data sobre suicidios
data_suicidios.rename(columns={
    'RegionCode': 'CodigoRegion',
    'RegionName': 'NombreRegion',
    'CountryCode': 'CodigoPais',
    'CountryName': 'NombrePais',
    'Year': 'Year',
    'Sex': 'Sexo',
    'AgeGroup': 'GrupoEdad',
    'Generation': 'Generacion',
    'SuicideCount': 'ConteoSuicidios',
    'CauseSpecificDeathPercentage': 'PorcentajeMuertesCausaEspecifica',
    'DeathRatePer100K': 'TasaMortalidadPor100K',
    'Population': 'Poblacion',
    'GDP': 'PIB',
    'GDPPerCapita': 'PIBPerCapita',
    'GrossNationalIncome': 'IngresoNacionalBruto',
    'GNIPerCapita': 'INBPerCapita',
    'InflationRate': 'TasaInflacion',
    'EmploymentPopulationRatio': 'RelacionEmpleoPoblacion'
}, inplace=True)
data_suicidios.head()
```

```python
# Fusionar los dos conjuntos de datos en base al país y año
data_combinada = pd.merge(data_suicidios, uso_internet, how='left', left_on=['NombrePais', 'Year'], right_on=['Pais', 'Year'])

# Ver las primeras filas del conjunto de datos combinado
data_combinada.head()
```

```python
data_combinada.columns
```

### Estimaciones para la cantidad de usuarios
Vamos a estimar la cantidad de usuarios que tenemos por grupo de edad, para esto habrá que apoyarnos de lo que vendría ser diversas estadísticas externas.

Según diferentes medios como lo son [Forbes](https://www.forbes.com/home-improvement/internet/internet-statistics/), [Statista](https://www.statista.com/topics/2576/us-millennials-internet-usage-and-online-shopping/#statisticChapter), [PewResearch](https://www.pewresearch.org/short-reads/2019/09/09/us-generations-technology-use/#:~:text=Almost%20all%20Millennials%20(nearly%20100,with%20just%2062%25%20of%20Silents.), podemos sacar una aproximación como la que tenemos a continuación.

```python
print("Grupos de edad en el DataFrame:")
print(data_combinada['GrupoEdad'].unique())

# Revisa si hay valores nulos en la columna 'GrupoEdad'
print("\nValores nulos en la columna 'GrupoEdad':")
print(data_combinada['GrupoEdad'].isnull().sum())
```

```python
# Definir un diccionario con los porcentajes de uso de internet por grupo de edad

# Son datos estáticos de el porcentaje de usuarios de esas edades con acceso a internet
ajuste_uso_internet = {
    '0-14 years': 0.50,
    '15-24 years': 0.75,
    '25-34 years': 0.90,
    '35-44 years': 0.85,
    '45-54 years': 0.80,
    '55-74 years': 0.54,
    '75+ years': 0.30,
    'Unknowm' : 0.00
}

# Ajustar los valores de uso de Internet en el conjunto de datos combinado
def ajustar_internet(fila):
    grupo_edad = fila['GrupoEdad']
    uso_base = fila['PoblacionUsandoInternetPorcentaje']
    return uso_base * ajuste_uso_internet.get(grupo_edad, 1)

data_combinada['UsoInternetAjustado'] = data_combinada.apply(ajustar_internet, axis=1)

# Ver las primeras filas del conjunto de datos ajustado
data_combinada[data_combinada['PoblacionUsandoInternetPorcentaje'] > 0][['NombrePais', 'Year', 'GrupoEdad', 'PoblacionUsandoInternetPorcentaje', 'UsoInternetAjustado']].head()
```

```python
# Definir la distribución estándar por edad y género
distribucion_estandar = {
    'GrupoEdad': ['0-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years', 'Unknown'],
    'PorcentajeDistribucion': [9.0, 6, 6, 13, 13, 1.0, 0.0]
}

df_distribucion = pd.DataFrame(distribucion_estandar)

# Convertir porcentajes a proporciones
df_distribucion['Proporcion'] = df_distribucion['PorcentajeDistribucion'] / 100


df_distribucion.head(6)
```

```python
data_combinada = pd.merge(data_combinada, df_distribucion, how='left', left_on=['GrupoEdad'], right_on=['GrupoEdad'])

data_combinada.head()
```

```python
# Asegurarse de que el porcentaje de uso de internet no exceda 100%
data_combinada['PoblacionUsandoInternetPorcentaje'] = data_combinada['PoblacionUsandoInternetPorcentaje'].clip(upper=1.0)
data_combinada['UsoInternetAjustado'] = data_combinada['UsoInternetAjustado'].clip(upper=1.0)


# Realiza el cálculo, divide el resultado entre  por los géneros y redondea a cero decimales
data_combinada['CantidadUsuariosInternet'] = ((
    data_combinada['Poblacion'] * 
    data_combinada['UsoInternetAjustado'] * 
    data_combinada['PoblacionUsandoInternetPorcentaje']*
    data_combinada['Proporcion'])
).round(0)
```

```python
data_combinada.columns
```

```python
# Seleccionar las columnas necesarias
columnas_necesarias = ['NombreRegion','NombrePais', 'CodigoPais', 'Year', 
'Sexo', 'GrupoEdad', 'TasaMortalidadPor100K','Poblacion',
'PoblacionUsandoInternetPorcentaje', 'CantidadUsuariosInternet',
'ConteoSuicidios'
]

# Crear un nuevo DataFrame con las columnas seleccionadas
data_final = data_combinada[columnas_necesarias]

# Ver las primeras filas del conjunto final
data_final.head()
```

## Análisis de los datos

```python
# Agrupar por año y calcular la media de las tasas de mortalidad por 100K
tendencias_temporales = data_final_limpia.groupby('Year').agg({
    'TasaMortalidadPor100K': 'mean',
    'ConteoSuicidios': 'sum',
    'CantidadUsuariosInternet': 'sum'
}).reset_index()

tendencias_temporales.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='TasaMortalidadPor100K', data=tendencias_temporales, marker='o', color='red')
plt.title('Promedio de Tasa de Mortalidad por Suicidio a lo Largo del Tiempo - Global')
plt.xlabel('Año')
plt.ylabel('Tasa de Mortalidad por 100K')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
# Gráfico de líneas usando Seaborn
sns.lineplot(
    data=data_final_limpia, 
    x='Year', 
    y='ConteoSuicidios', 
    hue='GrupoEdad',
    marker='o',    
)

plt.title('Número Total de Suicidios a lo Largo del Tiempo - Global')
plt.xlabel('Año')
plt.ylabel('Conteo de Suicidios')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='CantidadUsuariosInternet', data=tendencias_temporales, marker='o', color='green')
plt.title('Número Total de Usuarios de Internet a lo Largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Cantidad de Usuarios de Internet')
plt.grid(True)
plt.show()
```

```python
# Correlación entre Cantidad de Suicidios y Cantidad de Usuarios de Internet
correlation_suicidios_internet = data_final_limpia[['ConteoSuicidios', 'Poblacion']].corr().iloc[0, 1]
print(f'Correlación entre Cantidad de Suicidios y Poblacion: {correlation_suicidios_internet}')
```

Podriamos asumir que la llegada del internet ayudó a bajar la tasa de mortalidad por suicidio, ya que para el 2000 donde el internet ya era algo que se metía a la vida de la gente la tasa empezó a bajar cuando en los años anteriores tenía una lista relativamente recta y estable.

```python
# Relación entre Tasa de Suicidios y Porcentaje de Internet Usado
plt.figure(figsize=(10, 6))
sns.scatterplot(x='CantidadUsuariosInternet', y='ConteoSuicidios', data=data_final_limpia, hue='Sexo', palette='viridis')
plt.title('Relación entre Cantidad de usuarios de internet y el Conteo de Suicidios - Por sexo')
plt.xlabel('Cantidad de usuarios de internet')
plt.ylabel('Conteo de suicidios')
plt.legend(title='Sexo')
plt.grid(True)
plt.show()
```

Podemos observar en este gráfico que en el caso de los hombres la cantidad de suicidio mantiene más o menos una relación positiva con la cantidad de usuarios de internet.

Podemos ver que mientras más gente se ha reducico pero aún así ha habido puntos donde el internet ya es bastante popular y la cantidad de suicidios es alta.

```python
# Relación entre Tasa de Suicidios y Porcentaje de Internet Usado
plt.figure(figsize=(10, 6))
sns.scatterplot(x='CantidadUsuariosInternet', y='ConteoSuicidios', data=data_final_limpia, hue='Year', palette='viridis')
plt.title('Relación entre Cantidad de usuarios de internet y el Conteo de Suicidios - Por años')
plt.xlabel('Cantidad de usuarios de internet')
plt.ylabel('Conteo de suicidios')
plt.legend(title='Sexo')
plt.grid(True)
plt.show()
```

A medida que aumenta la cantidad de usuarios de Internet la tasa de mortalidad por suicidio disminuye, podría indicar una relación negativa.

```python
# Convertir 'Sexo' a variable numérica: Female=0, Male=1
data_final_limpia['Sexo_Num'] = data_final_limpia['Sexo'].map({'Female': 0, 'Male': 1})

# Convertir 'GrupoEdad' a una variable numérica usando codificación ordinal
# Aquí usamos un mapeo simple basado en el orden de los grupos de edad
grupo_edad_map = {
    '0-14 years': 0,
    '15-24 years': 1,
    '25-34 years': 2,
    '35-54 years': 3,
    '55-74 years': 4,
    '75+ years': 5
}
data_final_limpia['GrupoEdad_Num'] = data_final_limpia['GrupoEdad'].map(grupo_edad_map)

# Seleccionar las columnas necesarias para la matriz de correlación
data_correlacion = data_final_limpia[['GrupoEdad_Num', 'CantidadUsuariosInternet', 'ConteoSuicidios', 'Poblacion', 'Sexo_Num']]

# Calcular la matriz de correlación
matriz_correlacion = data_correlacion.corr()

matriz_correlacion
```

## Conclusión

### ¿Afecta realmente? Sí, el internet afecta de manera negativa al suicidio.

Con la llegada de internet, se puede observar un declive notable en las tasas de suicidio. Esto sugiere que personas que antes podían sentirse desconectadas o sin propósito han encontrado en internet un sentido para vivir. Internet ofrece una amplia gama de oportunidades y recursos que pueden influir positivamente en la vida de las personas.

Aunque se ha debatido extensamente sobre los peligros de internet, como la negatividad y el ciberacoso, estos aspectos negativos parecen ser superados por los beneficios que ofrece. Internet proporciona acceso a trabajo, amistades, amor y pasiones, lo cual tiene un impacto significativo en la calidad de vida y el bienestar emocional.

Sin embargo, es importante considerar estos resultados con cautela. Los datos utilizados no están específicamente curados para tratar un tema tan delicado como el suicidio, además de ser un tema tabú, lo que puede afectar la fiabilidad de los datos. Además, se utilizaron estimaciones fijas para obtener estos resultados, lo que puede introducir ciertas limitaciones en la precisión de las conclusiones.
